#Diagram

Man kan åskådliggöra information på ett enkelt sätt med hjälp av olika diagram. Det finns olika typer diagram som kan visa olika typer av information. Vilket diagram man väljer beror på vad man vill visa. Några av de vanligaste diagram är **cirkeldiagram**, **linjediagram**, **stapeldiagram**, **stolpdiagram** och **histogram**.

##Stapeldiagram:

Stapeldiagram visar frekvensen på hur ofta en viss händelse förekommer.

**Exempel 1**

I en skola har eleverna skapat en tabell över vilket godis som är deras favorit. Eftersom det de undersöker inte är numeriska värden (godis kan inte beskrivas med tal), och det de vill visa är hur många elever som gillar varje sorts godis (frekvensen), så använda de ett  **stapeldiagram**.


Favorit godis  | Antal elever
-------------------|------------------
Chocklad       | 30
Kola           | 14
Lakrits        | 9
Karameller     | 26
Tuggumi        | 21




```python
# @title Stapeldiagram, Exempel 1 Favorit Godis
import numpy as np
from matplotlib import pyplot as plt


# ändra värde här

# x-koordinat, vågräta axlen
godis = ["Choklad", "Kola", "Lakrits", "Karameller", "Tuggummi"]
# y-koordinat, heights of bars
antal = [30, 14, 9, 26, 21]


plt.bar(godis, antal, width = 0.5)
plt.xlabel('olika sorters godis')
plt.grid()
plt.ylabel('antalet elever som gillar')
plt.show()
```

På den vågräta axeln brukar man redovisa de olika händelse som har undersökts. På den lordräta axeln redovisar man förekomsten eller frekvensen av händelsen genom stapelns höjd.

## Cirkeldiagram

Cirkeldiagram används för att redovisa hur delar förhåller sig till helheten. Man delar upp den hela i cirkelsektorer (tårtbitar) där varje sektor representeras med en färg och storleken på sektorn anges i procent.


Om vi istället är intresserade av att visa andelen elever som gillar varje sorts godis, är det lämpligt med ett **cirkeldiagram**.





```python
# @title Cirkeldiagram, Exempel 1 Favorit Godis

# ingeting behöver ändra här

plt.pie(antal, labels = godis, autopct='%1.1f%%')
plt.show()
```

## Stolpdiagram:

Stolpdiagram används i liknande syfte som stapeldiagram. Skillanden är att här på den vågräta axeln redovisar man en (diskret) variabel, d.v.s. det som undersöks är tal/gå att beskrivas med tal.

**Exempel 2**

I samma skolan har eleverna också gjort en tabell över hur många godis bitar eleverna åt under veckan. Eftersom det som undersöks är tal (antal gånger, och det är en variabel) och vi är intresserad av antal frekvensen, så kan vi använder oss av ett **stodiagram**.

Godis bitar | Antal elever
-------------------|------------------
0           | 12
1           | 26
2           | 40
3           | 16
4           | 6
5           | 0




```python
# @title Stolpdiagram, Exempel 2 Antal Godisbitar i Veckan

# x-koordinat, vågräta axlen
godis_bitar = [0, 1, 2, 3, 4, 5]
# y-koordinat
antal_elever = [12, 26, 40, 16, 6, 0]

plt.stem(godis_bitar, antal_elever)
plt.xlabel('antal godis')
plt.grid()
plt.ylabel('antal elever')
plt.show()
```

###Skilland Mellan Stapeldiagram och Stolpdiagram:

För att sammanfatta:

*   Stolpdiagram används när det som undersöks kan beskrivas med tal
*   Stapeldiagram används när det som undersöks kan inte beskrivas med tal
*   I ett stapeldiagram kan man ändra ordningen på elementen längs den vågräta axeln, medan detta inte är möjligt i ett stolpdiagram





```python
# @title Samma Stapeldiagram, Exempel 1 Favorit Godis
import numpy as np
from matplotlib import pyplot as plt

# x-koordinat, vågräta axlen
godis = ["Choklad", "Kola", "Lakrits", "Karameller", "Tuggummi"]
# y-koordinat, heights of bars
antal = [30, 14, 9, 26, 21]

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].bar([godis[1], godis[0]]+[i for i in godis[1:]], [antal[1], antal[0]]+[i for i in antal[1:]], width = 0.5,
           color=['tab:blue', 'tab:red', 'tab:orange', 'tab:blue', 'tab:blue'])
axs[1].bar(godis, antal, width = 0.5,
           color=['tab:red', 'tab:orange', 'tab:blue', 'tab:blue', 'tab:blue'])


plt.show()
```

## Histogram:

När det som ska representeras innehåller många olika värden blir ett stolpdiagram svårt att avläsa. Man kan dela in värdena i flera klasser/katagorier/intervaller, och dessa klasser ritas som staplar intill varandra längs den vågräta axeln. Ett sådant digram kallas ett histogram.

**Exempel 3**

Nu har eleverna också gjort en undersökning om "*hur mycket pengar spenderar du på godis varje vecka*". Undersökningen gav följande resultat:


```python
# @title Undersökningsresultat, Exempel 3 Pengar Spenderat på Godis

# create data
import numpy as np
np.random.seed(10)
s = list(np.abs(np.random.normal(65, 33, 100)))
s[s.index(max(s))] = 160
s[s.index(min(s))] = 0
print([round(i) for i in s])
```


kr spenderat på godis | Antal elever

120 - 140          | 3
140 - 160          | 3



Eftersom vi har många värden i olika klasser, kan vi dela upp dem i olika klasser och sedan använda ett histogram:

```python

# @title Histogram, Exempel 3 Pengar Spenderat på Godis

klasser = 8



plt.hist(s, klasser, color="green")
plt.axis([0,160,0,35])
plt.xlabel("pengar (kr) spenderat på godis")
plt.ylabel("antal elever")
plt.grid()
plt.show()
```

## Linjediagram:

Linjediagram används för att illustrera hur något förändras med tiden. På den vågräta axeln brukar man ange förändringen i tid och på den lordräta axeln ange värdet vid motsvarande tidpunkten.

**Exempel 4**

Eleverna gjorde en undersökning på choklad konsumption per person under senaste år. Vi vill visa förändring i tid och därför gör vi en **linjediagram**.


Chocklad Konsumption | År
---------------|------------------
12.1           | 2015
12.4           | 2016
13.8           | 2017
13.1           | 2018
15.0           | 2019
15.2           | 2020
14.9           | 2021
15.4           | 2022
15.9           | 2023

```python
# @title Linjediagram, Exempel 4 Chocklad Konsumption per Person

ar  = np.arange(2015,2024)
konsumption  = [12.1, 12.4, 13.8, 13.1, 15.0, 15.2, 14.9, 15.4, 15.9]

plt.plot(ar, konsumption,'-o')
plt.xlabel("år")
plt.ylabel("chocklad konsumption")
plt.grid()
plt.show()
```

## Sammanfattning

Illustrera förändring över tid: **Linjediagram**

Illustrera andelen i procent: **Cirkeldiagram**

Illustrera förekomsten eller frekvens:

> Om det som undersöks är tal: **Stolpdiagram**

> Om det som undersöks inte är tal: **Stapeldiagram**

> Om det som undersöks kan delas i olika klasser: **Histogram**

# Vilseledande Diagram

Vilseledande diagram är diagram eller grafer som försöker få dig att tro något som egentligen inte är sant. Det kan göras genom att t.ex. ändra skalorna på axlarna, använda vissa färger eller former om kan ge en överdriven bild av informationen.

```python
# @title Vilseledande Linjediagram

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].plot(ar, konsumption,'o-')
axs[0].axis([2015,2023,0,50])

axs[1].semilogy(ar, konsumption,'o-')
axs[1].axis([2015,2023,12,16])

plt.show()
```
